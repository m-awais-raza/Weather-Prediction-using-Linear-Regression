```python
i = df[((df.precipMM==0)&(df.cloudcover==0))].index # since there are many zeroes in rain therefore to drop some of them we used any condition
```

```python
df.drop(i, inplace=True)
```

```python
import datetime as dt
df['date_time'] = pd.to_datetime(df['date_time'])
df['date_time'] = df['date_time'].map(dt.datetime.toordinal)
```

```python
# moonrise and moonset contains strings of no moonrise and no moonset which we will now replace with zero
df['moonrise'].replace('No moonrise',0,inplace=True)
df['moonset'].replace('No moonset',0,inplace=True)
```

```python
df['moonset'] = pd.to_datetime(df['moonset'])
df['moonset'] = df['moonset'].map(dt.datetime.toordinal)
```

```python
df['moonrise'] = pd.to_datetime(df['moonrise'])
df['moonrise'] = df['moonrise'].map(dt.datetime.toordinal)
```

```python
df = df.drop(df[["sunrise","sunset","uvIndex.1"]],axis=1)
```

```python
df = df.drop(df[["maxtempC"]],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df,label,test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
np.round(X_train_scaled.describe(), 1) # mean =0 and std=1 shows that standard scalar has been executed successfully
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train_scaled,y_train)
```

```python
# prediction on Training data
y_pred = lr.predict(X_test_scaled)
```

```python
# variance score: 1 means perfect prediction
#print('Variance score: {}'.format(lr.score(X_test_scaled, y_train)))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
